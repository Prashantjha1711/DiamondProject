```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df.iloc[:,:-1]
calc_vif(X)
```

```python
df.shape
```

Dropping variables should be an iterative process starting with the variable having the largest VIF value because other variables highly capture its trend. If you do this, you will notice that VIF values for other variables would have reduced, too, although to a varying extent.

```python
df.drop(['color_G'],axis=1,inplace = True)
```

## Applying Models

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr_train = lr.predict(x_train)
lr_test = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
print('Train Mape Linear reg : ',mean_absolute_percentage_error(y_train,lr_train))
```

```python
print('Test Mape Linear reg : ',mean_absolute_percentage_error(y_test,lr_test))
```

```python
from sklearn.tree import DecisionTreeRegressor 
classifier = DecisionTreeRegressor()
classifier.fit(x_train,y_train)
```

```python
dt_train = classifier.predict(x_train)
dt_test = classifier.predict(x_test)
```

```python
print('Train Mape Decision tree: ',mean_absolute_percentage_error(y_train,dt_train))
```

```python
print('Test Mape Decision tree : ',mean_absolute_percentage_error(y_test,dt_test))
```

```python
# pickling the model
import pickle
pickle_out = open("classifier.pkl", "wb")
pickle.dump(classifier, pickle_out)
pickle_out.close()
```
